```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import numpy as np
```

```python
# Damped harmonic oscillator
# x'' + 2zwx' + w^2x = 0

z = 0.002
w = 2.0
A = 3.0

w_d = w*np.sqrt(1-z**2)
k = w*z

xs = []
vs = []

a = A

fig, ax = plt.subplots()
ax.set_xlim(-4, 4)
ax.set_ylim(-6, 6)

graph, = ax.plot([], [])

def init():
    graph.set_data([], [])
    return graph,

def update(frame):
    t = frame/10.0
    global a
    a = a*np.exp(-k)
    x = a*np.cos(w_d*t)
    v = a*w_d*np.sin(w_d*t)
    xs.append(x)
    vs.append(v)
    graph.set_data(xs, vs)
    return graph,

animate = FuncAnimation(fig, update, frames=997, interval=20, init_func=init, blit=True)

plt.close()

HTML(animate.to_html5_video())
```
